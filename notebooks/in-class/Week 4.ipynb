```python
# open and write to files

import wikipedia


def get_page(page_title):
    # Returns text of wikipedia page as a dictionary
    try:
        page = wikipedia.page(search_result)
        return { 'title': page.title, 'text': page.content }
    except:
        return { 'title': '', 'text': '' }


# gets list of painters from wikipedia
search_results = wikipedia.search('Piracy')

result_pages = []

# get full data for each result as dictionary and append to list
for search_result in search_results:
    print('Checking ' + search_result)
    search_dict = get_page(search_result)
    result_pages.append(search_dict)

for result_page in result_pages:
    print('Writing ' + result_page['title'])
    write_file = open(result_page['title'] + '.txt', 'w+')
    write_file.write(result_page['text'])
    write_file.close()
```
